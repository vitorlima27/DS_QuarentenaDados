```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%precision %.2f
pd.options.display.float_format = '{:,.2f}'.format

#uri = "https://github.com/guilhermesilveira/enem-2018/blob/master/MICRODADOS_ENEM_2018_SAMPLE_43278.csv?raw=true"
dados = pd.read_csv('datasets/sample_enem2018.csv')
dados.head()
```

```python
dados.describe()
```

```python
colunas_de_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']
dados_notas = dados[colunas_de_notas].dropna()
dados_notas.columns = ['ciencias_naturais', 'ciencias_humanas', 'linguagem_codigo', 'matematica', 'redacao']
dados_notas.head()
```

```python
dados_notas['ciencias_naturais'].fillna(dados_notas['ciencias_naturais'].mean(), inplace = True)
dados_notas['ciencias_humanas'].fillna(dados_notas['ciencias_humanas'].mean(), inplace = True)
dados_notas['linguagem_codigo'].fillna(dados_notas['linguagem_codigo'].mean(), inplace = True)
dados_notas['matematica'].fillna(dados_notas['matematica'].mean(), inplace = True)
dados_notas['redacao'].fillna(dados_notas['redacao'].mean(), inplace = True)
```

```python
dados_notas['ciencias_naturais'].replace(0.,np.NAN)
dados_notas['ciencias_humanas'].replace(0.,np.NAN)
dados_notas['linguagem_codigo'].replace(0.,np.NAN)
dados_notas['matematica'].replace(0.,np.NAN)
```

```python
#Usando a função dropna, removi todos os valores nan das colunas de notas
dados_notas.dropna(subset=['ciencias_naturais', 'ciencias_humanas', 'linguagem_codigo', 'matematica'], inplace=True)
dados_notas.head()
```

```python
x_4_notas = dados_notas[['ciencias_naturais', 'ciencias_humanas', 'matematica', 'redacao']]
x_4_notas.head()
```

```python
y_adivinhar = dados_notas['linguagem_codigo']
y_adivinhar
```

```python
#Divisão dos dados em dados de treino e dados de teste
from sklearn.model_selection import train_test_split
import numpy as np

np.random.seed(43267)

# f(x) = y
x_treino, x_teste, y_treino, y_teste = train_test_split(x_4_notas, y_adivinhar)

print(x_treino.shape)
print(x_teste.shape)
print(y_treino.shape)
print(y_teste.shape)
```

```python
x_treino.head()
```

```python
#Utilizando o modelo LinearSVR
from sklearn.svm import LinearSVR

modelo = LinearSVR()
modelo.fit(x_treino, y_treino)
```

```python
predicoes_notas_linguagem = modelo.predict(x_teste)
predicoes_notas_linguagem[:5]
```

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x=y_teste.values, y=predicoes_notas_linguagem)
```

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x=y_teste.values, y=y_teste.values - predicoes_notas_linguagem)
```

```python
#Predição feita na aula

plt.figure(figsize=(9,9))
sns.scatterplot(x=y_teste.values, y=y_teste - x_teste.mean(axis=1))
```

```python
#Predição feita na aula
plt.figure(figsize=(9,9))
sns.scatterplot(x=y_teste.values, y=y_teste - y_treino.mean())
```

```python
#Utilizando o modelo de árvore de regressão

from sklearn.tree import DecisionTreeRegressor

modelo = DecisionTreeRegressor()
modelo.fit(x_treino, y_treino)
predicoes_notas_linguagem = modelo.predict(x_teste)
plt.figure(figsize=(9,9))
sns.scatterplot(x=y_teste.values, y=predicoes_notas_linguagem)
```

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x=x_teste['matematica'].values, y=predicoes_notas_linguagem)
sns.scatterplot(x=x_teste['matematica'].values, y=y_teste.values)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_teste, predicoes_notas_linguagem)
```

```python
from sklearn.dummy import DummyRegressor

modelo_dummy = DummyRegressor()
modelo_dummy.fit(x_treino, y_treino)
dummy_predicoes = modelo_dummy.predict(x_teste)

mean_squared_error(y_teste, dummy_predicoes)
```
